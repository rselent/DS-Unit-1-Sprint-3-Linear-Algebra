## 1.1 Create a two-dimensional vector and plot it on a graph

```
import math
import matplotlib.pyplot as plt
import numpy as np

aVector = [ 1.2, 1.6]

plt.arrow( .1, .2, aVector[0], aVector[1])
plt.xlim( -.2,2)
plt.ylim( -.2,2)
plt.show();
```

## 1.2 Create a three-dimensional vecor and plot it on a graph

```
from mpl_toolkits.mplot3d import Axes3D

oneVectorButThreeDee = [ 1.2, 1.6, .94]

threeDeeVectors = np.array( [[.1, .1, .1, 1.2, 1.6, .94]])

eye, dont, like, three, dee, vectors = zip(*threeDeeVectors)

fig = plt.figure()
ax = Axes3D( fig)
ax.quiver( eye, dont, like, three, dee, vectors)
ax.set_xlim( [0, 2])
ax.set_ylim( [0, 2])
ax.set_zlim( [0, 2])
ax.set_xlabel( 'x')
ax.set_ylabel( 'y')
ax.set_zlabel( 'z')
plt.show()
```
